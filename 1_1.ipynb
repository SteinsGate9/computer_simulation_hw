First we use Oula method to produce inverse of (a,M)
which indicates a*inverse(a,M) mod M = 1

```python
# functions to compute a^-1
# gcd function
def gcd(a, b):
    k = a // b
    remain = a % b
    while remain != 0:
        a = b 
        b = remain
        k = a // b
        remain = a % b
    return b

# compute inverse
def _halfdone(a, b):
    if b == 0:
        return 1, 0
    else:
        k = a // b
        remainder = a % b
        x1, y1 = _halfdone(b, remainder)
        x, y = y1, x1 - k * y1
        return x, y

def inverse(a,m):
    if gcd(a,m) != 1:
        return None
    else:
        x, y = _halfdone(a,m)
        return x % m

```

Setting up for the job, A B X0 M are parameters generated by random function.

```python
# settings for A
import numpy as np

# random initialization
np.random.seed(100)
A = np.random.randint(100)
X0 = np.random.randint(100)
M = np.random.randint(100)
B = inverse(A, M)

# if there is no inverse 
if B == None:
    print("no inverse for B")

# zip parameters into dictions  
params={'A':A, 'X0':X0, 'M':M, 'B':B}
print(A,X0,M,B)
```

test the linear model for A

```python
# test for A
def linear_mod(steps, **params):
    listt = [params['X0']]
    A = params['A']
    M = params['M']
    for i in range(steps):
        listt.append((A*listt[-1])%M)
    return listt

# print a length 100 list 
count = 0 
for i in linear_mod(100,**params):
    print(i, end=' ')
    if count != 10:
        count = count + 1
    else:
        print('\n')
        count = 0
```
